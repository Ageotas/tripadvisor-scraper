```python
import csv
import sys
from pprint import pprint
from topic_extraction import score_keyphrases_by_textrank
from collections import defaultdict

def main(csv_file):
    corpus = defaultdict(str)
    
    with open(csv_file, 'r', encoding='utf-8') as csvfile:
        reader = csv.reader(csvfile)
        header = reader.__next__()
        for row in reader:
            row_data = dict(zip(header, row))
            corpus[row_data['rating']] += row_data['title'] + " " + row_data['body']
    
    analysis = {}
    for k, v in corpus.items():
        analysis[int(float(k))] = score_keyphrases_by_textrank(v)
        
    pprint(analysis)
```

Run the above function to create a ranked database of keyphrases with correlation scores

```python
import pandas as pd

main("reviewsCleaned.csv")
```
